```python
import pandas as pd
from scipy.stats import mode
```

```python
# Input
data = pd.read_csv('BPI_Challenge_2012-training.csv')
```

```python
# Assign position number to each event
df_sort = data.set_index(data.groupby('case concept:name').cumcount(), append = True)
df_sort = df_sort.reset_index()
df_sort = df_sort[["level_1", "event concept:name"]]

# Find the most frequent event in each position using mode
df_new = df_sort.groupby('level_1')['event concept:name'].apply(lambda x: mode(x)[0][0]).reset_index()
df_new = df_new.rename(columns={'level_1':'position', 'event concept:name':'Predicted event'})
df_new
```
